```python
# import libraries
import numpy as np
import pandas as pd
import ggplot as gp
from ggplot import *
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from IPython.core.display import display, HTML
color = sns.color_palette()
plt.style.use('ggplot')
# display(HTML("<style>.container { width:100% !important; }</style>"))
abalone = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data", \
    header=None, names = ["Sex", "Length", "Diameter", "Height", "Whole_Weight", "Shucked_Weight", "Viscera_Weight",\
    "Shell_Weight","Rings"])
```

```python
abalone.head()
```

```python
abalone = abalone[abalone.Sex != 'I']
abalone.head()
abalone.count()
```

```python
abalone.corr()
```

```python
plt.subplots(figsize=(15,10))
sns.heatmap(abalone.corr(), xticklabels=abalone.columns.values, yticklabels=abalone.columns.values, annot=True, cmap="YlGnBu")
plt.show()
```

```python
abalone.groupby('Sex').size()
```

```python
g = sns.FacetGrid(abalone, col="Sex", size=8, aspect=.9)
g.map(plt.hist, "Length")
plt.show()
```
