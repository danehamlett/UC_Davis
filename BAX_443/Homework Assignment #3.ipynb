## Assignment I: BoW and TF-IDF

### Part I

#### Use Amazon book reviews (text documents) dataset (small dataset with 10 reviews).

```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cross_validation import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Import Data & Preview Results
df = pd.read_csv("https://raw.githubusercontent.com/danehamlett/UC_Davis/master/BAX_443/Small-Book%20Reviews%20from%20Amazon.csv", 
                sep=",", names=['Rating','Review'])
df.head()
```

### Part II

#### Use Bag of Words (BoW) and TF-IDF (CountVectorizer and TfidfVectorizer in scikit-learn).

##### Bag of Words (BoW)

```python
# Create a Corpus
corpus = df['Review'].tolist()

# Fitting the CountVectorizer In Order To:
#  1) Tokenize
#  2) Build the Vocabulary
vect = CountVectorizer()

# Fit the Corpus and Assign List Indices
vect.fit(corpus)

# Determine the Size of the Corpus
print(str(len(vect.vocabulary_)) + ' elements in the Corpus.' + '\n')

# Print the Corpus
print(vect.vocabulary_)
```

```python
# Create a Bag of Words Representation of the Data in a Document-Term Matrix
bag_of_words = vect.transform(corpus)

# Preview the Matrix
print(bag_of_words.toarray())
```

```python
# Define the Feature Names
feature_names = vect.get_feature_names()

# Query the Document-Term Matrix to Identify the Index of a Sample Word
print(vect.vocabulary_.get("sale"))

# Query the Document-Term Matrix to Identify the Sample Word of an Index
print(feature_names[681])
```

```python
# Confirm Number of Features
number_of_features = len(feature_names)
print(str(number_of_features) + ' features')
```

##### TF-IDF (Term Frequency - Inverse Document Frequency)

```python
# Fitting the CountVectorizer In Order To:
#  1) Tokenize
#  2) Build the Vocabulary
tfidf_vect = TfidfVectorizer()

# Fit the Corpus and Assign List Indices
tfidf_vect.fit(corpus)

# Determine the Size of the Corpus
print(str(len(tfidf_vect.vocabulary_)) + ' elements in the Corpus.' + '\n')

# Print the Corpus
print(tfidf_vect.vocabulary_)
```

```python
# Create a Bag of Words Representation of the Data in a Document-Term Matrix
tfidf = tfidf_vect.transform(corpus)

# Preview the Matrix
print(tfidf.toarray())
```

```python
# Define the Feature Names
tfidf_feature_names = tfidf_vect.get_feature_names()

# Query the Document-Term Matrix to Identify the Index of a Sample Word
print(tfidf_vect.vocabulary_.get("think"))

# Query the Document-Term Matrix to Identify the Sample Word of an Index
print(tfidf_feature_names[792])
```

```python
# Confirm Number of Features
tfidf_number_of_features = len(tfidf_feature_names)
print(str(tfidf_number_of_features) + ' features.')
```

#### Try unigram and bigram parameters and observe their effect on number of features.

```python
# Define Ngram Range
ngram_start = 1
ngram_end = 5

# Define Ngram-baed Vectorizer & Fit the Model
ngram_vect = CountVectorizer(ngram_range=(ngram_start, ngram_end))
ngram_vect.fit(corpus)
ngram_bag_of_words = ngram_vect.transform(corpus)
ngram_feature_names = ngram_vect.get_feature_names()

# Print the Results
ngram_number_of_features = len(ngram_feature_names)
print(str(ngram_number_of_features) + ' features when capturing words with a length between ' + str(ngram_start)
     + ' and ' + str(ngram_end) + '.' + '\n')
print('The number of features will increase as the ngram range widens.' + '\n')

# Sample a Value
print(str(ngram_vect.vocabulary_.get("quite painful when left alone")) + '\n')

# Confirm Index
print(ngram_feature_names[6030] + '\n')

# List the Full Vocabulary
print(ngram_vect.vocabulary_)
```
